#### utils

```python
## utils
def dense_plot(nrow, ncol, fontsize, n_bins, df, title, figsize = (16,8), df2 = None, duo = False):
        fig, axes = plt.subplots(nrow, ncol, figsize = figsize)
        plt.suptitle(title, fontsize = fontsize, y = 1.0)

        for i, col in enumerate(df.columns):
            ax = axes[i//ncol][i%ncol]
            if duo:
                sns.histplot(df[col], bins = n_bins, color = '#7b5aa6',
                           alpha = 0.8, edgecolor = 'white', ax = ax, label = 'raw')
                sns.histplot(df2[col], bins = n_bins,
                           color = 'darkgray', alpha = 0.3,
                           edgecolor = 'white', ax = ax, label = 'processed')
                ax.legend()

            else:
                sns.histplot(df[col], bins = n_bins, color = '#7b5aa6', alpha = 0.8,
                       edgecolor = 'white', ax = ax)
                ax.set_xlabel(f'{col}')

        for j in range(i+1, nrow * ncol):
            axes[j//ncol][j%ncol].axis('off')

        plt.tight_layout()
```

```python
## QQ plot
def qq_plot(nrow, ncol, fontsize, df, title, figsize = (16,8)):
    fig, axes = plt.subplots(nrow, ncol, figsize =figsize)
    plt.suptitle(title, fontsize = fontsize, y = 1.0)

    for i, col in enumerate(df.columns):
        ax = axes[i//ncol][i%ncol]
        qqplot = sm.qqplot(data = df[col], dist = norm, loc = df[col].mean(),
                                    scale = df[col].std(),
                           markerfacecolor='#7b5aa6', markeredgecolor= 'None',
                           markersize = 5.5, alpha = 0.3, ax = ax)
        sm.qqline(line='45', color = 'darkgray', lw = 1.5, ax = ax)

        ax.set_ylabel(f'{col}')


    for j in range(i+1, nrow * ncol):
        axes[j//ncol][j%ncol].axis('off')

    plt.tight_layout()
```

```python
def model_selection(model, X, y, path = '', model_label = '',
                    param_dict = None, scoring = None):
    try:
        clf = load(f'{path}/{model_label}.joblib')
    except:
        clf = GridSearchCV(estimator =model,
                        param_grid = param_dict, scoring = scoring, n_jobs = -1,
                    cv = 3, refit = 'AUC', verbose = True, return_train_score = True)

        clf.fit(X, y)
        ## save model
        dump(clf, f'{path}/{model_label}.joblib')

    print(clf.best_estimator_)
    print('=='*20)
    print(f'auc score = {clf.score(X, y):.4f}') # roc_auc
    print(f'auprc score = {average_precision_score(y, clf.predict_proba(X)[:,1]):.4f}')
    print(f'accuracy = {accuracy_score(y, clf.predict(X)):.4f}')
    print('=='*20)
    return clf

def cf_matrix(clf, X, y, label = ''):
    cf_mat = confusion_matrix(clf.predict(X), y)
    fig, ax = plt.subplots(figsize = (6,4))
    texts = [[f'True Negative = {cf_mat[0,0]}',f'False Positive = {cf_mat[0,1]}'], \
            [f'False Negative = {cf_mat[1,0]}', f'True Positive = {cf_mat[1,1]}']]
    sns.heatmap(cf_mat, cmap = 'Purples', alpha = 0.7, annot = np.array(texts), fmt = '',\
                ax = ax, linewidths = .1, linecolor = 'white')
    ax.set_title('Confusion matrix of {label}', fontsize = 12, y = 1.05)
    ax.set_xticklabels(labels = [r'O:E < 1', 'O:E $\geq$ 1'], )
    ax.tick_params(length = .1)
    ax.set_yticklabels(labels = [r'O:E < 1', 'O:E $\geq$ 1'], rotation = 0)
    

def roc_auc_curve(y_true, clf, X, label = ''):
    y_pred_prob = clf.predict_proba(X)[:,1]
    y_pred = clf.predict(X)                              
    fpr, tpr, threshold = roc_curve(y_true, y_pred)
    fig, ax = plt.subplots(1,1, figsize = (6,4))
    sns.lineplot(fpr, tpr, color =  violet, alpha = .8, ax = ax, label = 'roc')

    ax.set_title(f'ROC curve: {label}', fontsize = 12, y = 1.05)
    ax.set_xlabel('False Positive Rate')
    ax.set_ylabel('True Positive Rate')
    ax.fill_between(fpr, tpr, color = violet, alpha = 0.3)
    ax.text(.5, .5, f'auc = {roc_auc_score(y_true, y_pred_prob):.4f}')

```

```python
path = os.getcwd()

# reduced_dim = pd.read_csv(f'{path}/IM_SNA_reduced_dimensions_11032023.csv', index_col = 0)
streamed = pd.read_csv(f'{path}/SNA_streamlined_mean_11032023 (1).csv', index_col = 0)
```

```python
## train-test split
streamed = streamed.groupby(streamed.index).mean()
dep = ['O:E'] # let's only focus on O:E
y =  streamed[dep]
# recode y into binary
y = np.where(y < 1, 0, 1)
X =  streamed.drop(dep + ['Encounter_index'], axis = 1)

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state= random_state)
print(f'Training set length = {len(X_train)}\nValidation set length = {len(X_test)}')
```

```python
## visualize distribution of whole network SNA metric
# already reduced dim
# size to transitivity
X_train_whole_SNA = X_train.iloc[:, np.argmax([i.startswith('size') for i in X_train.columns]): \
    np.argmax([i.startswith('transitivity') for i in X_train.columns]) + 1]

# no missing values
dense_plot(3, 4, 15, 12, X_train_whole_SNA, 'Density plot for whole network SNA metric')
print(X_train_whole_SNA.isnull().sum())
```

```python
## visualize skewness & KS-test
qq_plot(3, 4, 15, X_train_whole_SNA, 'QQ-plot for whole network SNA metric', figsize = (18,12))

# Also Kolmogorov-Smirnov test for goodness of fit
# https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.kstest.html
threshold = 0.05
skewed = [i for i in X_train_whole_SNA.columns if kstest(np.array(
    X_train_whole_SNA[i][~np.isnan(X_train_whole_SNA[i])]), norm(loc = X_train_whole_SNA[i].mean(),
                                                             scale = X_train_whole_SNA[i].std()).cdf)[1] < threshold]
print(f'Number of skewed features = {len(skewed)}\nNumber of features = {X_train_whole_SNA.shape[1]}')
skewed_df = pd.DataFrame(skewed, columns = ['skewed features'])
print(skewed_df)
```

##### **3. normalize, log-transform**
---

```python
## normalize features with < 0 values
scaler = MinMaxScaler()
X_train_neg = X_train_whole_SNA[[col for col in skewed if X_train_whole_SNA[col].min() < 0]]
X_train_neg_scaled = scaler.fit_transform(X_train_neg)
X_train_neg_scaled = pd.DataFrame(X_train_neg_scaled, columns=X_train_neg.columns, index = X_train.index)

# map(negative, positive) -> (0,1) in training set
X_train_whole_SNA[X_train_neg_scaled.columns] = X_train_neg_scaled
```

```python
## log transform
X_train_whole_SNA_log = pd.DataFrame(np.log(X_train_whole_SNA[skewed] + 1e-10), columns = skewed,
                                     index = X_train_whole_SNA.index)
X_train_whole_SNA_log = pd.concat([X_train_whole_SNA_log,
                                   X_train_whole_SNA.loc[:,~X_train_whole_SNA.columns.isin(skewed)]], axis = 1)
X_train_whole_SNA_log = X_train_whole_SNA_log[X_train_whole_SNA.columns]

# visualize
# check for normality

dense_plot(3, 4, 15, 12, X_train_whole_SNA_log, 'Density plot for whole network SNA metric')
```

```python
## drop outliers
# since sna metrics seem to be driven by extreme values. I use wider quantiles vs. Q1, Q3
Q_l = X_train_whole_SNA_log.apply(lambda x: x.quantile(0.005))
Q_r = X_train_whole_SNA_log.apply(lambda x: x.quantile(0.995))
X_train_whole_SNA_trunc = (X_train_whole_SNA_log >= Q_l) & (X_train_whole_SNA_log <= Q_r)
X_train_whole_SNA_drop = X_train_whole_SNA_log[X_train_whole_SNA_trunc].dropna(axis = 0)
```

###### 1. fill missing values
---
Fill with 0 & mode/median in raw space first

(i) Since not missing at random, and individual SNA metrics tend to be bounded by \[0,1\], try to fill in missing values in the raw space first.\
(a) count-based or reversed distance based: fill with 0 to indicate lack of interaction.\
          -  indegree, outdegree, authority, hub (ref: https://www.jstor.org/stable/20453505), transitivity,  betweeness(# shortest paths/random)?\
(b) distance-based: fill with mode/median.\
        - closeness (1/avg distance), eccentricity centrality (1/max dist), harmonic(1/sum(distance)), coreness
\
(ii) Also, forgo log transform temporarily due to undesired results

```python
# only select individual SNA metrics
X_train_ind_SNA = X_train.iloc[:, np.argmax([i.startswith('transitivity') for i in X_train.columns]) + 1: -1]

# many missing values
missing_val = X_train_ind_SNA.isnull().sum()
missing_val = missing_val.sort_values(ascending = False)
print(missing_val)

# check for normality per role
roles = ['Medicine Attending', 'Medicine Intern', 'Medicine Resident', 'Medicine APP', 'Nurse', 'SW']
for role in roles:
    X_train_sub = X_train_ind_SNA.iloc[:, [s.startswith(role) for s in X_train_ind_SNA.columns]]
    sub_title = 'Density plot for individual SNA metric ' + '(' + role + ')'
    dense_plot(3, 4, 15, 12, X_train_sub, sub_title, figsize = (16, 8))
```

```python
# separate individual SNA into count/reversed distance-based metrics vs. distance based.
count_based = r'(.*indegree|.*outdegree|.*authority|.*hub|.*transitivity|.*betweeness)'
X_train_ind_SNA_count = X_train_ind_SNA.iloc[:, [len(re.findall(count_based, i))>0 for i in X_train_ind_SNA]]
X_train_ind_SNA_distance = X_train_ind_SNA.loc[:, ~X_train_ind_SNA.columns.isin(X_train_ind_SNA_count.columns)]

# fillna with 0 and mode/median, respectively
X_train_ind_SNA_count = X_train_ind_SNA_count.fillna(0).set_index(X_train_ind_SNA.index)
X_train_ind_SNA_distance = X_train_ind_SNA_distance.fillna(X_train_ind_SNA_distance.median()).set_index(X_train_ind_SNA.index)

X_train_ind_SNA_filled = pd.concat([X_train_ind_SNA_count, X_train_ind_SNA_distance], axis = 1)[X_train_ind_SNA.columns]
# X_train_ind_SNA_filled.head()
X_train_ind_SNA_filled.isnull().sum()
```

---

```python
## preprocess test whole SNA

X_test_whole_SNA = X_test.iloc[:, np.argmax([i.startswith('size') for i in X_test.columns]): \
    np.argmax([i.startswith('transitivity') for i in X_test.columns]) + 1]

scaler = MinMaxScaler()
X_test_neg = X_test_whole_SNA[[col for col in skewed if X_test_whole_SNA[col].min() < 0]]
X_test_neg_scaled = scaler.fit_transform(X_test_neg)
X_test_neg_scaled = pd.DataFrame(X_test_neg_scaled, columns=X_test_neg.columns, index = X_test_whole_SNA.index)

# map(negative, positive) -> (0,1) in training set
X_test_whole_SNA[X_test_neg_scaled.columns] = X_test_neg_scaled

## log transform
X_test_whole_SNA_log = pd.DataFrame(np.log(X_test_whole_SNA[skewed]), columns = skewed,
                                    index = X_test_whole_SNA.index)
X_test_whole_SNA_log = pd.concat([X_test_whole_SNA_log,
                                   X_test_whole_SNA.loc[:,~X_test_whole_SNA.columns.isin(skewed)]], axis = 1)
X_test_whole_SNA_log = X_test_whole_SNA_log[X_test_whole_SNA.columns]

## drop outliers
# since sna metrics seem to be driven by extreme values. I use wider quantiles vs. Q1, Q3
Q_l = X_test_whole_SNA_log.apply(lambda x: x.quantile(0.005))
Q_r = X_test_whole_SNA_log.apply(lambda x: x.quantile(0.995))
X_test_whole_SNA_trunc = (X_test_whole_SNA_log >= Q_l) & (X_test_whole_SNA_log <= Q_r)
X_test_whole_SNA_drop = X_test_whole_SNA_log[X_test_whole_SNA_trunc].dropna(axis = 0)

## dim reduction
n_PC = 3
X_test_scaled = StandardScaler().fit_transform(X_test_whole_SNA_drop)
pca = PCA(n_components = n_PC)
X_test_processed = pca.fit_transform(X_test_scaled)

## transform
X_test_whole_SNA_ = pd.DataFrame(X_test_processed[:,:n_PC],
                                 columns = [f'whole_PC_{i}' for i in range(1, n_PC+1)])
X_test_whole_SNA_.set_index(X_test_whole_SNA_drop.index, inplace = True)
# X_test_whole_SNA_
```

```python
## preprocess val ind SNA
X_test_ind_SNA = X_test.iloc[:, np.argmax([i.startswith('transitivity') for i in X_test.columns]) + 1: -1]

# separate individual SNA into count/reversed distance-based metrics vs. distance based.
count_based = r'(.*indegree|.*outdegree|.*authority|.*hub|.*transitivity|.*betweeness)'
X_test_ind_SNA_count = X_test_ind_SNA.iloc[:, [len(re.findall(count_based, i))>0 for i in X_test_ind_SNA]]
X_test_ind_SNA_distance = X_test_ind_SNA.loc[:, ~X_test_ind_SNA.columns.isin(X_test_ind_SNA_count.columns)]

# fillna with 0 and mode/median, respectively
X_test_ind_SNA_count = X_test_ind_SNA_count.fillna(0).set_index(X_test_ind_SNA.index)
X_test_ind_SNA_distance = X_test_ind_SNA_distance.fillna(X_test_ind_SNA_distance.median()).set_index(X_test_ind_SNA.index)

X_test_ind_SNA_filled = pd.concat([X_test_ind_SNA_count, X_test_ind_SNA_distance], axis = 1)[X_test_ind_SNA.columns]
```

###### reduce dimension

```python
#PCA
n_PC = 13
X_test_scaled = StandardScaler().fit_transform(X_test_ind_SNA_filled)
pca = PCA(n_components= n_PC)
X_test_processed = pca.fit_transform(X_test_scaled)

X_test_ind_SNA_ = pd.DataFrame(X_test_processed[:,:n_PC],
                                 columns = [f'ind_PC_{i}' for i in range(1, n_PC+1)])
X_test_ind_SNA_.set_index(X_test_ind_SNA_filled.index, inplace = True)
X_test_ind_SNA_
```

```python
## combine
X_test_SNA_ = X_test_whole_SNA_.join(X_test_ind_SNA_, how = 'inner')
X_test_SNA_.shape
```
